```python
from pyspark.sql.functions import (dayofmonth,dayofyear,hour,
                                   month,year,weekofyear,format_number,
                                   date_format)
```

#### Generally using pyspark.sql.function we use df.select(func_name(col_on_which_it_is_called)).show()

```python
df.select(dayofmonth(df['Date'])).show()
```

```python
df.select(hour(df['Date'])).show()
```

```python
df.select(month(df['Date'])).show()
```

##### Average closing price per year

```python
df.withColumn("Year",year(df['Date'])).show()
```

```python
newdf = df.withColumn("Year",year(df['Date']))
newdf
```

#### mean changes the column name

```python
newdf.groupBy('Year').mean().show()
```

```python
newdf.groupBy('Year').mean().select(['Year','avg(Close)']).show()
```

```python
result = newdf.groupBy('Year').mean().select(['Year','avg(Close)'])
```

```python
new = result.withColumnRenamed("avg(Close)","Average_Closing_Price")
new.show()
```

```python
new.select('Year',format_number(new['Average_Closing_Price'],2).alias('Avg Close')).show()
```
